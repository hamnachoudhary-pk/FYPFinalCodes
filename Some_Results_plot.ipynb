```python
# PP-YOLO After fine tune 


import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

# Simulated true labels and predictions after fine-tuning
true_labels = np.array(
    ["Business Jet"]*360 + 
    ["Cargo Airplane"]*200 + 
    ["Container Trolley Type A"]*1200 + 
    ["Person Wearing Reflective Vests"]*250 + 
    ["Standard Car"]*320 + 
    ["Transport Container"]*1280
)

# Predictions with small error rates introduced
pred_labels = np.array(
    ["Business Jet"]*355 + ["Cargo Airplane"]*5 +
    ["Cargo Airplane"]*195 + ["Business Jet"]*5 +
    ["Container Trolley Type A"]*1180 + ["Standard Car"]*20 +
    ["Person Wearing Reflective Vests"]*240 + ["Standard Car"]*10 +
    ["Standard Car"]*310 + ["Cargo Airplane"]*10 +
    ["Transport Container"]*1270 + ["Container Trolley Type A"]*10
)

# Confusion Matrix
cm = confusion_matrix(true_labels, pred_labels, labels=[
    "Business Jet", "Cargo Airplane", "Container Trolley Type A",
    "Person Wearing Reflective Vests", "Standard Car", "Transport Container"
])

# Plot confusion matrix
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", 
            xticklabels=["Business Jet", "Cargo Airplane", "Container Trolley Type A",
                         "Person Wearing Reflective Vests", "Standard Car", "Transport Container"],
            yticklabels=["Business Jet", "Cargo Airplane", "Container Trolley Type A",
                         "Person Wearing Reflective Vests", "Standard Car", "Transport Container"])
plt.title("Expected Confusion Matrix After Fine-Tuning")
plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.show()

# Classification Report
report = classification_report(true_labels, pred_labels, target_names=[
    "Business Jet", "Cargo Airplane", "Container Trolley Type A",
    "Person Wearing Reflective Vests", "Standard Car", "Transport Container"
])

print("Expected Classification Report After Fine-Tuning:\n")
print(report)

# Simulated Training Loss curve (after fine-tuning)
steps = np.arange(0, 2000)
loss = 1.5 + 0.5*np.exp(-steps/300) + 0.05*np.random.randn(len(steps))

plt.figure(figsize=(10,5))
plt.plot(steps, loss, label="Training Loss", color="blue")
plt.title("Expected Training Loss Curve After Fine-Tuning")
plt.xlabel("Training Steps")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set a seed for reproducibility (optional)
np.random.seed(42)

# Base true positives (diagonal values)
base_tps = [360, 365, 440, 220, 350, 465]
total_tp = sum(base_tps)  # 2,200

# Total ground truths based on recall (TP / 0.6652  3,285)
total_ground_truths = int(total_tp / 0.66524499654769)  # ~3,285
total_fn = total_ground_truths - total_tp  # ~1,085

# Initialize confusion matrix with base true positives
confusion_matrix = np.diag(base_tps).astype(float)  # Use float to avoid type issues

# Class labels
classes = ['Airside Safety Vehicle', 'Ambulance Vehicle', 'Business Jet', 
           'Cargo Airplane', 'Cleaning Vehicle', 'Common Aircraft']

# Generate random false positives (sum to ~120 to match precision ~0.9489)
# Total predictions = TP + FP  2,200 + 120 = 2,320
# Precision = TP / (TP + FP)  2,200 / 2,320  0.9483
fp_total = 120
fp_distribution = np.random.randint(0, 15, size=(6, 6))  # Random ints 0-14
fp_distribution = fp_distribution - np.diag(np.diag(fp_distribution))  # Zero diagonal
current_fp_sum = np.sum(fp_distribution)
fp_adjustment = int((fp_total - current_fp_sum) / 6)  # Round to integer
remaining_fp = fp_total - current_fp_sum - (fp_adjustment * 6)  # Remaining to distribute

# Adjust fp_distribution with integer adjustment
for i in range(6):
    fp_distribution[i, i] = fp_adjustment  # Distribute evenly along diagonal (temp storage)
fp_distribution = fp_distribution - np.diag([fp_adjustment] * 6)  # Remove from diagonal
if remaining_fp > 0:
    indices = np.random.choice(30, remaining_fp, replace=False)  # 30 off-diagonal positions
    rows, cols = divmod(indices, 6)
    for r, c in zip(rows, cols):
        if r != c:  # Ensure off-diagonal
            fp_distribution[r, c] += 1

# Add false positives to the confusion matrix (off-diagonal only)
for i in range(6):
    for j in range(6):
        if i != j:
            confusion_matrix[i, j] = fp_distribution[i, j]

# Convert to integer for final display
confusion_matrix = confusion_matrix.astype(int)

# Create the plot
plt.figure(figsize=(10, 8))  # Larger figure size for better readability
im = plt.imshow(confusion_matrix, interpolation='nearest', cmap='Blues')
plt.colorbar(label='Count per image')

# Set ticks and labels
plt.xticks(np.arange(len(classes)), classes, rotation=45, ha='right', fontsize=12)
plt.yticks(np.arange(len(classes)), classes, fontsize=12)

# Add values in cells
for i in range(len(classes)):
    for j in range(len(classes)):
        plt.text(j, i, confusion_matrix[i, j], ha='center', va='center', color='black', fontsize=12)

# Title and labels
plt.title('Confusion Matrix (Diagonal shows True Positives per image)', fontsize=14)
plt.ylabel('True Labels', fontsize=12)
plt.xlabel('Predicted Labels', fontsize=12)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```

# Faster RCNN plot 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import pandas as pd

# Define the class labels
class_labels = [
    'Business Jet',
    'Cargo Airplane',
    'Container Trolley Type A',
    'Personnel Wearing Reflective Vests',
    'Standard Car',
    'Transport Container'
]

# Correct class sample sizes to be consistent and non-zero
# This assumes the user wants to have samples for all classes.
# Note: The original sizes were '91', '114', '64', '102', '120'.
# 'Business Jet' had 0 samples initially.
class_sizes = [
    2,  # Corrected to have samples
    91, # From original prompt
    114,  # From original prompt
    64, # From original prompt
    102, # From original prompt
    120  # Added for 'Transport Container' to match labels
]

# Ensure the number of labels matches the number of class sizes
if len(class_labels) != len(class_sizes):
    print("Error: The number of class labels does not match the number of sample sizes.")
else:
    # Generate perfect classification data
    true_labels = []
    pred_labels = []

    for i, size in enumerate(class_sizes):
        true_labels.extend([i] * size)
        pred_labels.extend([i] * size)

    # Create confusion matrix
    cm = confusion_matrix(true_labels, pred_labels)

    # Create the plot
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm,
                annot=True,
                fmt='d',
                cmap='Blues',
                xticklabels=class_labels,
                yticklabels=class_labels,
                cbar_kws={'label': 'Count'},
                annot_kws={'size': 14, 'weight': 'bold'})

    plt.title('Confusion Matrix (Diagonal shows True Positives per class)',
              fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('Predicted Label', fontsize=14, fontweight='bold')
    plt.ylabel('True Label', fontsize=14, fontweight='bold')

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)

    # Adjust layout to prevent label cutoff
    plt.tight_layout()

    # Display the matrix values
    print("Enhanced Confusion Matrix:")
    print("=" * 50)

    # The loop will now work correctly as class_labels and cm have the same size.
    for i, label in enumerate(class_labels):
        print(f"{label}: {cm[i][i]} samples")

    print(f"\nTotal samples: {np.sum(cm)}")
    print(f"Accuracy: {np.trace(cm) / np.sum(cm) * 100:.1f}%")

    # Calculate additional metrics
    print("\nPer-class Statistics:")
    print("=" * 50)
    for i, label in enumerate(class_labels):
        # Handle cases where denominator is zero to avoid division by zero error
        precision = cm[i][i] / np.sum(cm[:, i]) if np.sum(cm[:, i]) > 0 else 0
        recall = cm[i][i] / np.sum(cm[i, :]) if np.sum(cm[i, :]) > 0 else 0
        print(f"{label}:")
        print(f"  Samples: {cm[i][i]}")
        print(f"  Precision: {precision:.3f}")
        print(f"  Recall: {recall:.3f}")
        print()

    plt.show()

    # Optional: Save the confusion matrix as a DataFrame for further analysis
    cm_df = pd.DataFrame(cm, index=class_labels, columns=class_labels)
    print("Confusion Matrix as DataFrame:")
    print(cm_df)
```
